```python
import tensorflow as tf
```

```python
sent = ['the glass of milk' , 
         ' the glass of juice' , 
        'the cup of tea ' , 
        'im a good boy' , 
        'im a good developer' , 
        'understand the meaning of words' ,
        'your videos are good']
       
       
       
       
                  
```

```python
sent
```

```python
### vocabulary size 

voc_size = 10000
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

### one-hot encoding

```python
onehot = [one_hot(words , voc_size) for words in sent ]

# here iterating through each and every words in sentence
# using one_hot and passing words and voc_size as parameter
# storing into one_hot for every word of each sentence where  ever the word is present denoted by 1 , that index we're extracting 

print(onehot)
```

### word embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
## pre padding 

sent_length = 8 # considering the maximum length of sentence to be  8 

embedded_docs = pad_sequences(onehot , padding = 'pre' , maxlen = sent_length)

## pad sequences will make sure that all the vectors of these words make it  8
## padding = 'pre' meaning we are adding zeros in the begining to make the size 8 
## padding ='post' meaning we are adding zeros in the last to size 8
print(embedded_docs)
```

```python
dim = 10 # feature representation (dimension) size = 10

model = Sequential()

model.add(Embedding(voc_size , 10 ,input_length=sent_length ))

## embedding layer is going to work similar to word2vec this is going to train all the neural networks

model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs[0])
```

```python
print(model.predict(embedded_docs))
```
