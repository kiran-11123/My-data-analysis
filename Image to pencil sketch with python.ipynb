```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
# get the image file
image='d:bird.jpeg'
org_image=cv2.imread(image)
```

```python
# display the original image
plt.imshow(org_image)
plt.axis('off')
plt.title('original image')
plt.show()
```

```python
# convert the image to greyscale
gray_image=cv2.cvtColor(org_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
plt.axis('off')
plt.title('gray image')
plt.show()
```

```python
# invert the grayscale image

inverted_gray_image=cv2.bitwise_not(gray_image)
```

```python
plt.imshow(inverted_gray_image)
plt.axis('off')
plt.title('inverted gray image')
plt.show()
```

```python
# blur the inverted image using the gaussainblur function
blurred_image=cv2.GaussianBlur(inverted_gray_image,(111,111),0)
```

```python
plt.imshow(blurred_image)
plt.axis('off')
plt.title('blurred image')
plt.show()
```

```python
# invert the blurred image

inverted_blur_image=cv2.bitwise_not(blurred_image)
```

```python
plt.imshow(inverted_blur_image)
plt.axis('off')
plt.title('inverted blur image')
plt.show()
```

```python
# create the pencil image by dividing the greyscale image by the inverting the image
pencil_sketch=cv2.divide(gray_image,inverted_blur_image,scale=256.0)
```

```python
plt.imshow(pencil_sketch)
plt.axis('off')

plt.show()
```

```python
# convert the pencil sketch to RGB
pencil_sketch_rgb=cv2.cvtColor(pencil_sketch,cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(pencil_sketch_rgb)
plt.axis('off')

plt.show()
```
